## 1. Khai báo thư viện & Tham số thực nghiệm

```python
# ── Thư viện ────────────────────────────────────────────
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
import warnings
from datetime import datetime
warnings.filterwarnings('ignore')
from datetime import datetime
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler, StandardScaler, LabelEncoder

# ── Tham số thực nghiệm ─────────────────────────────────
BASE_DIR   = os.path.abspath(os.path.join(os.getcwd(), '..')) \
             if os.path.basename(os.getcwd()) == 'prj' \
             else os.getcwd()

DATA_PATH  = os.path.join(BASE_DIR, 'data', 'iris.csv')
OUTPUT_DIR = os.path.join(BASE_DIR, 'exps_')

'''
Nội dung của các file lưu lại các lần chạy và chuẩn bị dữ liệu để train.
1. eda_log.xlsx: gồm 5 sheet:
                + Thong_tin_chung: lưu thông tin ngày chạy, số dòng.
                + Thong_ke_mo_ta: min, max, mean, std từng cột.
                + Missing_Values: số lượng và % gia trị thiếu từng cột.
                + Phan_phoi_lop: mỗi lớp có bao nhiêu mẫu, chiếm bao nhiêu %.
                + Train _Test_Split: đường dẫn 6 file csv, số mẫu train/test.
2. train_raw.csv: data train chưa chuẩn hóa.
3. test_raw.csv: data test chưa chuẩn hóa.
4. train_minmax.csv: data train đã chuẩn hóa minmax.
5. test_minmax.csv: data test đã chuẩn hóa minmax.
6. train_standard: data train đã chuẩn hóa Standard.
7. test_standard: data test đã chuẩn hóa Standard.
'''
EDA_LOG        = os.path.join(OUTPUT_DIR, 'eda_log.xlsx')
TRAIN_DATA_RAW = os.path.join(OUTPUT_DIR, 'train_raw.csv')
TEST_DATA_RAW  = os.path.join(OUTPUT_DIR, 'test_raw.csv')
TRAIN_DATA_MM  = os.path.join(OUTPUT_DIR, 'train_minmax.csv')
TEST_DATA_MM   = os.path.join(OUTPUT_DIR, 'test_minmax.csv')
TRAIN_DATA_STD = os.path.join(OUTPUT_DIR, 'train_standard.csv')
TEST_DATA_STD  = os.path.join(OUTPUT_DIR, 'test_standard.csv')
TEST_SIZE      = 0.3 # tỉ lệ để test model
RANDOM_STATE   = 42

# tên của mỗi cột trong data.csv
COL_NAMES      = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'class']

os.makedirs(OUTPUT_DIR, exist_ok=True)
sns.set_theme(style='whitegrid')
plt.rcParams['figure.dpi'] = 100

print('✅ Khai báo thư viện và tham số xong!')
print(f'   Thời gian chạy : {datetime.now().strftime("%Y-%m-%d %H:%M:%S")}')
print(f'   File dữ liệu   : {DATA_PATH}')
print(f'   Tỉ lệ test     : {TEST_SIZE}')
```

## 2. Đọc dữ liệu

```python
# Đọc file CSV (không có header)
df = pd.read_csv(DATA_PATH, header=None, names=COL_NAMES)

print(f'Shape: {df.shape}  ({df.shape[0]} dòng x {df.shape[1]} cột)')
print(f'Tên cột: {df.columns.tolist()}')
df.head(10)
```

## 3. Thông tin tổng quan

```python
print('=== THÔNG TIN DATASET ===')
df.info()
```

```python
print('=== THỐNG KÊ MÔ TẢ ===')
df.describe().round(4)
```

## 4. Kiểm tra Missing Values & Duplicates

```python
print('=== MISSING VALUES ===')
missing = df.isnull().sum()
missing_pct = (missing / len(df) * 100).round(2)
df_missing = pd.DataFrame({'Số thiếu': missing, '% thiếu': missing_pct})
print(df_missing)

print(f'\n=== DUPLICATES ===')
dup = df.duplicated().sum()
print(f'Số dòng trùng lặp: {dup}')
if dup > 0:
    print('⚠️  Có dòng trùng lặp!')
    print(df[df.duplicated()])
else:
    print('✅ Không có dòng trùng lặp')
```

## 5. Phân phối nhãn (Class Distribution)

```python
class_counts = df['class'].value_counts()
print('=== PHÂN PHỐI LỚP ===')
print(class_counts)

fig, axes = plt.subplots(1, 2, figsize=(12, 4))

# Bar chart
colors = ['#4C72B0', '#DD8452', '#55A868']
axes[0].bar(class_counts.index, class_counts.values, color=colors, edgecolor='black', linewidth=0.8)
axes[0].set_title('Số lượng mẫu theo lớp', fontsize=13, fontweight='bold')
axes[0].set_xlabel('Lớp')
axes[0].set_ylabel('Số lượng')
for i, v in enumerate(class_counts.values):
    axes[0].text(i, v + 0.5, str(v), ha='center', fontweight='bold')

# Pie chart
axes[1].pie(class_counts.values, labels=class_counts.index,
            autopct='%1.1f%%', colors=colors, startangle=90,
            wedgeprops={'edgecolor': 'white', 'linewidth': 2})
axes[1].set_title('Tỉ lệ phân phối lớp', fontsize=13, fontweight='bold')

plt.tight_layout()
plt.savefig(os.path.join(OUTPUT_DIR, 'plot_class_distribution.png'), bbox_inches='tight')
plt.show()
print('✅ Dữ liệu cân bằng: 50 mẫu mỗi lớp')
```

## 6. Phân phối từng feature (Histogram)

```python
features = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']
fig, axes = plt.subplots(2, 2, figsize=(12, 8))
axes = axes.flatten()

for i, feat in enumerate(features):
    for cls, color in zip(df['class'].unique(), colors):
        axes[i].hist(df[df['class'] == cls][feat], bins=15,
                     alpha=0.6, label=cls, color=color, edgecolor='white')
    axes[i].set_title(f'Phân phối: {feat}', fontweight='bold')
    axes[i].set_xlabel(feat)
    axes[i].set_ylabel('Tần suất')
    axes[i].legend(fontsize=8)

plt.suptitle('Phân phối các đặc trưng theo lớp', fontsize=14, fontweight='bold', y=1.01)
plt.tight_layout()
plt.savefig(os.path.join(OUTPUT_DIR, 'plot_histogram.png'), bbox_inches='tight')
plt.show()
```

## 7. Boxplot - Phát hiện Outlier

```python
fig, axes = plt.subplots(2, 2, figsize=(12, 8))
axes = axes.flatten()

for i, feat in enumerate(features):
    df.boxplot(column=feat, by='class', ax=axes[i],
               boxprops=dict(color='navy'),
               medianprops=dict(color='red', linewidth=2),
               whiskerprops=dict(color='navy'),
               capprops=dict(color='navy'),
               flierprops=dict(marker='o', color='orange', markersize=6))
    axes[i].set_title(f'Boxplot: {feat}', fontweight='bold')
    axes[i].set_xlabel('Lớp')
    axes[i].set_ylabel(feat)

plt.suptitle('Boxplot các đặc trưng theo lớp (phát hiện outlier)', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.savefig(os.path.join(OUTPUT_DIR, 'plot_boxplot.png'), bbox_inches='tight')
plt.show()
```

## 8. Correlation Heatmap

```python
fig, ax = plt.subplots(figsize=(8, 6))
corr = df[features].corr()
mask = np.triu(np.ones_like(corr, dtype=bool))

sns.heatmap(corr, annot=True, fmt='.2f', cmap='coolwarm',
            mask=mask, ax=ax, linewidths=0.5,
            annot_kws={'size': 12, 'weight': 'bold'},
            vmin=-1, vmax=1, center=0)
ax.set_title('Ma trận tương quan giữa các đặc trưng', fontsize=14, fontweight='bold', pad=15)

plt.tight_layout()
plt.savefig(os.path.join(OUTPUT_DIR, 'plot_correlation.png'), bbox_inches='tight')
plt.show()
print('\n📌 Nhận xét:')
print('  - petal_length và petal_width tương quan rất cao (≈0.96)')
print('  - sepal_width tương quan âm với các đặc trưng còn lại')
```

## 9. Pairplot - Quan hệ giữa các đặc trưng

```python
palette = {'Iris-setosa': '#4C72B0', 'Iris-versicolor': '#DD8452', 'Iris-virginica': '#55A868'}
g = sns.pairplot(df, hue='class', palette=palette,
                 diag_kind='kde', plot_kws={'alpha': 0.6})
g.fig.suptitle('Pairplot - Quan hệ giữa các đặc trưng', y=1.02, fontsize=14, fontweight='bold')
plt.savefig(os.path.join(OUTPUT_DIR, 'plot_pairplot.png'), bbox_inches='tight')
plt.show()
```

## 10. Violin Plot

```python
fig, axes = plt.subplots(2, 2, figsize=(13, 8))
axes = axes.flatten()

for i, feat in enumerate(features):
    sns.violinplot(data=df, x='class', y=feat, ax=axes[i],
                   palette=palette, inner='quartile', linewidth=1.2)
    axes[i].set_title(f'Violin: {feat}', fontweight='bold')
    axes[i].set_xlabel('')
    axes[i].tick_params(axis='x', rotation=10)

plt.suptitle('Violin Plot các đặc trưng theo lớp', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.savefig(os.path.join(OUTPUT_DIR, 'plot_violin.png'), bbox_inches='tight')
plt.show()
```

## 11. Chia dữ liệu Train/Test (7/3)

```python
print('=== CHIA DỮ LIỆU TRAIN / TEST (7/3) ===')

X = df[features]
y = df['class']

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=TEST_SIZE, random_state=RANDOM_STATE, stratify=y)

print(f'Tổng mẫu    : {len(df)}')
print(f'Train set   : {len(X_train)} mẫu ({len(X_train)/len(df)*100:.0f}%)')
print(f'Test set    : {len(X_test)} mẫu ({len(X_test)/len(df)*100:.0f}%)')
print(f'\nPhân phối train:\n{y_train.value_counts()}')
print(f'\nPhân phối test:\n{y_test.value_counts()}')
```

## 12. Chuẩn hóa dữ liệu & Lưu file

```python
def save_split(X_tr, X_te, y_tr, y_te, train_path, test_path, scaler=None, scaler_name='raw'):
    """Chuẩn hóa (nếu cần) và lưu file train/test"""
    X_tr_s, X_te_s = X_tr.copy(), X_te.copy()
    if scaler:
        X_tr_s = pd.DataFrame(scaler.fit_transform(X_tr), columns=features, index=X_tr.index)
        X_te_s = pd.DataFrame(scaler.transform(X_te),     columns=features, index=X_te.index)

    train_df = pd.concat([X_tr_s, y_tr], axis=1)
    test_df  = pd.concat([X_te_s, y_te], axis=1)
    train_df.to_csv(train_path, index=False)
    test_df.to_csv(test_path,  index=False)
    print(f'✅ [{scaler_name}] Train → {train_path}')
    print(f'✅ [{scaler_name}] Test  → {test_path}')
    return train_df, test_df

# Raw
train_raw, test_raw = save_split(X_train, X_test, y_train, y_test,
                                  TRAIN_DATA_RAW, TEST_DATA_RAW, scaler_name='raw')

# MinMax
train_mm, test_mm = save_split(X_train, X_test, y_train, y_test,
                                TRAIN_DATA_MM, TEST_DATA_MM,
                                scaler=MinMaxScaler(), scaler_name='minmax')

# Standard
train_std, test_std = save_split(X_train, X_test, y_train, y_test,
                                  TRAIN_DATA_STD, TEST_DATA_STD,
                                  scaler=StandardScaler(), scaler_name='standard')
```

## 13. So sánh phân phối trước/sau chuẩn hóa

```python
fig, axes = plt.subplots(3, 4, figsize=(16, 10))
titles = ['Raw (Gốc)', 'MinMax Scaled', 'Standard Scaled']
datasets = [train_raw, train_mm, train_std]

for row, (title, data) in enumerate(zip(titles, datasets)):
    for col, feat in enumerate(features):
        axes[row, col].hist(data[feat], bins=15, color=colors[row],
                            edgecolor='white', alpha=0.85)
        axes[row, col].set_title(f'{title}\n{feat}', fontsize=9, fontweight='bold')
        axes[row, col].set_ylabel('Tần suất' if col == 0 else '')

plt.suptitle('So sánh phân phối: Raw vs MinMax vs Standard', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.savefig(os.path.join(OUTPUT_DIR, 'plot_scaling_compare.png'), bbox_inches='tight')
plt.show()
```

## 14. Ghi kết quả EDA vào Excel

```python
#%% md
## 14. Ghi kết quả EDA vào Excel
#%%
# Timestamp để phân biệt từng lần chạy
timestamp = datetime.now().strftime('%Y-%m-%d %H:%M:%S')
run_id = datetime.now().strftime('%Y%m%d_%H%M%S')  # dùng đặt tên sheet

# ── Chuẩn bị dữ liệu ghi ────────────────────────────────
df_info = pd.DataFrame({
    'Thông tin': ['Thời gian chạy', 'File dữ liệu', 'Số dòng', 'Số cột',
                  'Số lớp', 'Missing values', 'Duplicates', 'Tỉ lệ split'],
    'Giá trị':   [timestamp, DATA_PATH,
                  df.shape[0], df.shape[1],
                  df['class'].nunique(), df.isnull().sum().sum(),
                  df.duplicated().sum(), f'Train {1-TEST_SIZE:.0%} / Test {TEST_SIZE:.0%}']
})

df_stats = df.describe().round(4).reset_index().rename(columns={'index': 'Thống kê'})

df_miss = pd.DataFrame({
    'Cột':      df.columns,
    'Số thiếu': df.isnull().sum().values,
    '% thiếu':  (df.isnull().sum() / len(df) * 100).round(2).values
})

df_class = df['class'].value_counts().reset_index()
df_class.columns = ['Lớp', 'Số lượng']
df_class['Tỉ lệ %'] = (df_class['Số lượng'] / len(df) * 100).round(2)

df_split = pd.DataFrame({
    'Tập':     ['Train (raw)', 'Test (raw)', 'Train (minmax)', 'Test (minmax)', 'Train (standard)', 'Test (standard)'],
    'File':    [TRAIN_DATA_RAW, TEST_DATA_RAW, TRAIN_DATA_MM, TEST_DATA_MM, TRAIN_DATA_STD, TEST_DATA_STD],
    'Số mẫu':  [len(X_train), len(X_test)] * 3,
})

# ── Ghi vào Excel ────────────────────────────────────────
import os
from openpyxl import load_workbook

# Kiểm tra file đã tồn tại chưa
file_exists = os.path.exists(EDA_LOG)

if not file_exists:
    # Lần đầu chạy → tạo file mới với sheet cố định + sheet lịch sử
    with pd.ExcelWriter(EDA_LOG, engine='openpyxl', mode='w') as writer:
        df_info.to_excel(writer,   sheet_name='Thong_tin_chung',  index=False)
        df_stats.to_excel(writer,  sheet_name='Thong_ke_mo_ta',   index=False)
        df_miss.to_excel(writer,   sheet_name='Missing_Values',   index=False)
        df_class.to_excel(writer,  sheet_name='Phan_phoi_lop',    index=False)
        df_split.to_excel(writer,  sheet_name='Train_Test_Split', index=False)
        # Sheet lịch sử — lần đầu
        df_info.to_excel(writer,   sheet_name=f'Run_{run_id}',    index=False)
    print(f'🆕 Tạo file mới: {EDA_LOG}')
else:
    # Lần sau → cập nhật sheet cố định + thêm sheet lịch sử mới
    with pd.ExcelWriter(EDA_LOG, engine='openpyxl', mode='a', if_sheet_exists='replace') as writer:
        df_info.to_excel(writer,   sheet_name='Thong_tin_chung',  index=False)
        df_stats.to_excel(writer,  sheet_name='Thong_ke_mo_ta',   index=False)
        df_miss.to_excel(writer,   sheet_name='Missing_Values',   index=False)
        df_class.to_excel(writer,  sheet_name='Phan_phoi_lop',    index=False)
        df_split.to_excel(writer,  sheet_name='Train_Test_Split', index=False)
        # Thêm sheet lịch sử mới, không ghi đè sheet cũ
        df_info.to_excel(writer,   sheet_name=f'Run_{run_id}',    index=False)
    print(f'➕ Đã thêm lịch sử vào: {EDA_LOG}')

print(f'✅ Sheet lịch sử lần này: Run_{run_id}')
print(f'\n📋 Cấu trúc file Excel:')
wb = load_workbook(EDA_LOG)
for i, sheet in enumerate(wb.sheetnames):
    tag = '← lần này' if sheet == f'Run_{run_id}' else ''
    print(f'   Sheet {i+1}: {sheet} {tag}')
wb.close()
```
